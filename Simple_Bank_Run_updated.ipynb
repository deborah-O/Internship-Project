```python
# Imports
import random
from matplotlib import pyplot as plt
```

```python
class Agent():
    def __init__ (self, patient, belief, money = 0):
        self.patience = patient
        self.money = money 
        self.beliefs = belief
    
```

```python
class Model():
    # Again, need to have "self" in the list of arguments
    # Also added N which will be the populate size
    def __init__(self, N , belief, max_iterations=100, total_reserves = 10000):
        # Need to initialise the model here
        self.N = N
        
        self.impatient_agents = list()
        self.reserves_history = list()
        self.patient_agents = list()
        
        # That would probably mean making a list of agents
        # As I've currently written it, all of the agents will initially be patient
        self.agents = list()
        for i in range(N):
            self.beliefs = (random.random()) 
            self.agents.append(Agent(True, belief = random.random()*0.5 + 0.5))
        
        # Make a variable to keep track of the number of impatient people
        impatient_agents = list() 
        for agent in self.agents:
            if agent.patience == False:
                impatient_agents.append(agent)
        self.num_of_impatients = len(impatient_agents)
        
        
        # Making variable for patient people
       
        patient_agents = list()
        for agent in self.agents:
            if agent.patience == True:
                patient_agents.append(agent)
        self.num_of_patients = len(patient_agents)
    
        
        # Keep a record of the model time
        self.time = 0
        
        # Note down the max iterations
        self.max_iterations = max_iterations
        self.total_reserves = total_reserves
        
        self.alpha1 = 0.008
        self.alpha2 = 0.02
        
   
    def run(self):
        
        for t in range(self.max_iterations):
            if random.random() > 0.5:
                    qual_info = (random.random())
                    print(qual_info)
            else:
                pass
            
            for agent in self.agents:
                if agent.patience == True:
                    if qual_info >= 0.5:
                        agent.beliefs -= self.alpha2
                    else: 
                        agent.beliefs -= self.alpha1
                else:
                    agent.beliefs -= self.alpha2
                
                if agent.patience == True:
                    if self.num_of_impatients > self.N/1.5 or agent.beliefs <= 0.5:
                        agent.patience = False
                 
                if agent.patience == False:
                    agent.money +=1
                    if self.total_reserves > 0:
                        self.total_reserves -= 1
                        
            self.reserves_history.append(self.total_reserves)
            
        
            # Update number of impatient people and model time
            self.time += 1
            # Make a variable to keep track of the number of impatient people
            impatient_agents = list()
            
            for agent in self.agents:
                if agent.patience == False:
                    impatient_agents.append(agent)
                    self.num_of_impatients = len(impatient_agents)
            self.impatient_agents.append(self.num_of_impatients)
            
            patient_agents = list()
            
            for agent in self.agents:
                if agent.patience == True:
                    patient_agents.append(agent)
                    self.num_of_patients = len(patient_agents)
            self.patient_agents.append(self.num_of_patients)
        
            print(self.time, self.num_of_impatients, self.num_of_patients)
    
```

```python
m = Model(100,0)
```

```python
for agent in m.agents:
    print(agent.patience, agent.beliefs, agent.money)
```

```python
m.run()
# for agent in m.agents:
timestep_no = range(len(m.impatient_agents))
plt.figure(figsize=(15,10))
plt.bar(timestep_no, m.impatient_agents, label = 'impatient')
plt.ylabel('Number of Impatient Agents')
plt.xlabel('Time Periods')
plt.title('Transition from Patient to Impatient Agents')
plt.legend()
plt.show()
```

```python
timestep_no = range(len(m.patient_agents))
plt.figure(figsize=(15,10))
plt.bar(timestep_no, m.patient_agents, label = 'patient')
plt.ylabel('Number of patient Agents')
plt.xlabel('Time Periods')
plt.title('Decrease in Patient Agents during Panic')
plt.legend()
plt.show()
```
